```python
def standardize_columns1(int):
    mean = int.mean()
    std_dev = int.std()
    standardized_columns1 = (int - mean) / std_dev
    return standardized_columns1
```

```python
for column in numerical_columns:
    df[column] = standardize_columns1(df[column])

df[numerical_columns.columns]
```

```python
def standardize_columns2(float):
    mean = float.mean()
    std_dev = float.std()
    standardized_columns2 = (float - mean) / std_dev
    return standardized_columns2
```

```python
for column in numerical_columns2:
    df[column] = standardize_columns2(df[column])

df[numerical_columns2.columns]
```
